#### Plot the correlation matrix using seaborn.

```python
# your code here
co=titanic.corr(method='pearson')
co
```

```python
sns.set(style='white')     # estilo blanco


mascara=np.triu(np.ones_like(co, dtype=bool))   # genera una mascara para tapar valores


cmap=sns.diverging_palette(0, 10, as_cmap=True)   # paleta de colores


sns.heatmap(co,# <-argumento
            mask=mascara,
            cmap=cmap,
            vmax=1,
            center=0,
            square=True,
            linewidth=0.5,
            cbar_kws={'shrink': 0.6},
            annot=True
           );
```
